# Spilting X and y data 

```python
X = data.drop(['price'], axis=1)
y = data['price']
```

```python
X.columns
```

```python
X_encoded = pd.get_dummies(X, columns=['model', 'transmission', 'fuelType'], dtype=int)
```

```python
X_encoded.head()
```

```python
X_encoded.columns
```

# Splitting test and train

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

# Correlation with pearsonr correlation

```python
from scipy.stats import pearsonr
```

```python
cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

correlations = {
    
        feature : pearsonr(X_encoded[feature], y)[0]
        for feature in cols
}

correlation_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Correlation with price'])
correlation_df.sort_values(by='Correlation with price', ascending=False)
```

# Mutual info regression


```python
X_encoded.columns
```

```python
features = ['year', 'mileage', 'tax', 'mpg', 'engineSize', 'model_ B-MAX',
       'model_ C-MAX', 'model_ EcoSport', 'model_ Edge', 'model_ Escort',
       'model_ Fiesta', 'model_ Focus', 'model_ Fusion', 'model_ Galaxy',
       'model_ Grand C-MAX', 'model_ Grand Tourneo Connect', 'model_ KA',
       'model_ Ka+', 'model_ Kuga', 'model_ Mondeo', 'model_ Mustang',
       'model_ Puma', 'model_ Ranger', 'model_ S-MAX', 'model_ Streetka',
       'model_ Tourneo Connect', 'model_ Tourneo Custom',
       'model_ Transit Tourneo', 'model_Focus', 'transmission_Automatic',
       'transmission_Manual', 'transmission_Semi-Auto', 'fuelType_Diesel',
       'fuelType_Electric', 'fuelType_Hybrid', 'fuelType_Other',
       'fuelType_Petrol']
```

```python
from sklearn.feature_selection import mutual_info_regression

mi_scores = mutual_info_regression(
    X_encoded[features],
    y,
    random_state=42
)

mi_df = pd.DataFrame({
    'Feature': features,
    'Mutual Information': mi_scores
}).sort_values(by='Mutual Information', ascending=False)

```

```python
mi_df = mi_df.sort_values(by='Mutual Information', ascending=False)
mi_df
```

```python
list(mi_df['Feature'])
```

Feature selection was performed using Mutual Information. However, after experimentation, the model achieved better performance when trained using all available features. Therefore, all features were retained for the final model.

**Feature Scalling**

```python
cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()

X_train[cols] = scalar.fit_transform(X_train[cols])
X_test[cols] = scalar.transform(X_test[cols])
```

```python
X_train
```

```python
X_test
```

# Linear Regression model 

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()

linear_model.fit(X_train, y_train)

```

```python
linear_model_prediction = linear_model.predict(X_test)
```

```python
comparison_df = pd.DataFrame({
    'Actual Price': y_test,
    'Predicted Price': linear_model_prediction
})

comparison_df
```

**Model Evaluation**

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score
```

```python
linear_model_r2 = r2_score(y_test, linear_model_prediction)

linear_model_adjusted_r2 = 1 - (1 - linear_model_r2) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)

linear_model_mse = mean_squared_error(y_test, linear_model_prediction)

linear_model_rmse = np.sqrt(linear_model_mse)

linear_model_mae = mean_absolute_error(y_test, linear_model_prediction)

print(f"r2_score : {linear_model_r2}")

print(f"Adjusted r2 : {linear_model_adjusted_r2}")

print(f"Mean Squared Error (MSE): {linear_model_mse}")

print(f"Root Mean Squared Error (RMSE): {linear_model_rmse}")

print(f"Mean Absolute Error (MAE): {linear_model_mae}")
```

```python
np.mean(y_test)
```

```python
np.mean(linear_model_prediction)
```

**linear_model_Prediction vs actual**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=linear_model_prediction)
plt.xlabel('Actual Price')
plt.ylabel('linear model Predicted Price')
plt.title('Actual vs linear_model')
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    'r--'
)
plt.show()
plt.savefig('images_linear_model/actual_vs_linear_model.png')
```

**Residuals difference graph of linear model**

```python
residuals = y_test - linear_model_prediction

plt.figure(figsize=(8, 6))
sns.histplot(residuals, bins=30, kde=True)
plt.title('Residual Distribution')
plt.xlabel('Residuals')
plt.show()
plt.savefig('images_linear_model/residual_distribution_linear_model.png')
```

# KNN-Regressor

```python
from sklearn.neighbors import KNeighborsRegressor

KNN_model = KNeighborsRegressor()
```

```python
KNN_prediction = KNN_model.predict(X_test)
```

**Model Evaluation**

```python
KNN_model_r2 = r2_score(y_test,KNN_prediction)

KNN_model_adjusted_r2 = 1 - (1 - KNN_model_r2) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)

KNN_model_mse = mean_squared_error(y_test, KNN_prediction)

KNN_model_rmse = np.sqrt(KNN_model_mse)

KNN_model_mae = mean_absolute_error(y_test, KNN_prediction)

print(f"r2_score : {KNN_model_r2}")

print(f"Adjusted r2 : {KNN_model_adjusted_r2}")

print(f"Mean Squared Error (MSE): {KNN_model_mse}")

print(f"Root Mean Squared Error (RMSE): {KNN_model_rmse}")

print(f"Mean Absolute Error (MAE): {KNN_model_mae}")
```
